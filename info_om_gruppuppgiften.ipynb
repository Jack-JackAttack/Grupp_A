# Gruppuppgift – E-handel: snabbrapport & GitHub-samarbete



Kort beskrivning (case)
Ni är ett analysteam på en fiktiv e-handelsplattform. Ledningen vill ha ett snabbt beslutsunderlag inför nästa kampanjperiod. Underlaget ska vara körbart och reproducerbart i GitHub och bestå av en kort, tydlig rapport-notebook som visar nyckeltal och figurer samt en kort README.

- Det här vill ledningen ha svar på
- Vad säljer? – vilka kategorier driver mest intäkt?
- Var säljer vi? – vilka städer står för störst intäkt?
- När säljer vi? – finns tidsmönster/säsong i försäljningen?
- Hur ser en typisk order ut? – AOV (Average Order Value) och spridning.
- Topp-listor – topp-3 kategorier efter intäkt.
- Eventuella avvikelser – något oväntat mönster som sticker ut?
- Ni ska avsluta rapporten med 2–3 korta rekommendationer (punktlista) baserat på era resultat, t.ex. “Satsa på kategori X i stad Y under månad Z”.

Dataset (CSV): ecommerce_sales.csv
Kolumner: order_id, date, city, category, price, units, revenue


Vad ska göras (krav för G)
Repo & samarbete
Sätt upp ett GitHub-repo för gruppen.
Arbeta via branches och Pull Requests.
Lägg all körbar analys i en Jupyter-notebook som fungerar från topp till botten.
Nyckeltal (minst 5)
Total intäkt och totalt antal enheter.
AOV (Average Order Value).
Intäkt per kategori.
Intäkt per stad.
Top-3 kategorier efter intäkt.
Visualiseringar (minst 2)
Exempel: intäkt per kategori (stapeldiagram), försäljning över tid (linje/vecka eller månad).
Tydliga rubriker, axlar och 1–2 meningar markdown som förklarar vad figuren visar.
Kodkvalitet
Minst 2 egna funktioner för beräkningar (t.ex. compute_revenue(...), top_categories(...)).
Notebooken ska köra utan fel från start till mål.
README
Kort vad ni gjort, hur man kör (ev. requirements) och roller/ansvar i teamet.
Kort slutsats (2–3 punkter):
 Lägg rubriken Rekommendationer överst eller nederst i notebooken.

notebooks/report.ipynb
README.md
